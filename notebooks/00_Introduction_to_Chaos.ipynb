```python
Markdown("""
## What is Chaos?

Chaos is a branch of mathematics and physics that studies how small differences in initial conditions can lead to vastly different outcomes.

**Key characteristics of chaotic systems:**
- Deterministic (no randomness involved)
- Highly sensitive to initial conditions
- Long-term behavior is unpredictable

This is often referred to as the **"butterfly effect"**.
""")
```

```python
Markdown("""
## Why Pendulums?

The **driven damped pendulum** is a simple yet powerful example of a chaotic system.

It includes:
- A restoring force (gravity)
- A damping force (friction or air resistance)
- An external periodic force (driving)

When these elements interact, they can produce extremely complex behaviorâ€”even from simple rules.

This makes pendulums an ideal system to:
- Visualize chaos
- Simulate using numerical methods
- Analyze with data science and machine learning
""")
```

```python
Markdown(r"""
## ðŸ§® Governing Equation

The equation of motion for the driven, damped pendulum is:

\[
\frac{d^2\theta}{dt^2} + q \frac{d\theta}{dt} + \sin(\theta) = A \cos(\omega t)
\]

Where:
- \( \theta \) is the angle
- \( q \) is the damping coefficient
- \( A \) is the driving force amplitude
- \( \omega \) is the driving frequency

We'll explore how this leads to chaos in later notebooks.
""")
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 20, 1000)
theta_simple = np.sin(t)
theta_chaotic = np.sin(t) + 0.3 * np.sin(3*t) + 0.2 * np.sin(7*t)

plt.figure(figsize=(12, 7))
plt.plot(t, theta_simple, label="Periodic motion")
plt.plot(t, theta_chaotic, label="Chaotic-like motion", alpha=0.7)
plt.title("Simple vs. Chaotic-like Pendulum Motion")
plt.xlabel("Time")
plt.ylabel("Theta (rad)")
plt.legend()
plt.grid(True)
plt.show()
```
